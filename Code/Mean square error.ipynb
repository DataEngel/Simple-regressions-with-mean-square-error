```python
# Import the libraries 
import numpy as np
from sklearn.datasets import load_boston
import matplotlib.pyplot as plt
```

```python
#Add the dataset 
boston = load_boston()

# Show the dataset
print(boston.DESCR) 
```

The formula to minimize the Mean Square Error (MSE): $\beta = (X^{T}X)^{-1}X^{T}Y$

```python
# We obtain the data and transform them into matrices
X = np.array(boston.data[:, 5])
Y = np.array(boston.target)

# Paint the points 
plt.scatter(X, Y, alpha=0.3)

# Add the column of 1 for the independent term and appy the transposed 
X = np.array([np.ones(506), X]).T

# Apply the formula MSE 
B = np.linalg.inv(X.T @ X) @ X.T @ Y

# We paint the regression line estimated by ordinary least squares 
plt.plot([4, 9], [B[0] + B[1] * 4, B[0] + B[1] * 9], c="green")

plt.xlabel('Average number of rooms per dwelling ')
plt.ylabel('y-Median value of owner-occupied homes in $1000 ')

plt.show()
```
